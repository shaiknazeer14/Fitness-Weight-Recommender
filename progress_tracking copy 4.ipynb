### Feature Engneering

```python
# First create a total reps column for each row
train_df['Total_Reps'] = train_df[['Set1_Reps', 'Set2_Reps', 'Set3_Reps']].mean(axis=1)

# Then group and average
train_df['Avg_reps_workout'] = train_df.groupby(
    ['day', 'month', 'year', 'Body_Part', 'Exercise_Name', 'Equipment_Type','Health condition']
)['Total_Reps'].transform('mean')
```

```python
# First create a total reps column for each row
test_df['Total_Reps'] = test_df[['Set1_Reps', 'Set2_Reps', 'Set3_Reps']].mean(axis=1)

# Then group and average
test_df['Avg_reps_workout'] = test_df.groupby(
    ['day', 'month', 'year', 'Body_Part', 'Exercise_Name', 'Equipment_Type','Health condition']
)['Total_Reps'].transform('mean')
```

```python
train_df.head()
```

```python
# Then group and average
train_df['Reps_diff_bw_sets'] = train_df.groupby(
    ['day', 'month', 'year', 'Body_Part', 'Exercise_Name', 'Equipment_Type',]
)['Total_Reps'].diff()
```

```python
# Then group and average
test_df['Reps_diff_bw_sets'] = test_df.groupby(
    ['day', 'month', 'year', 'Body_Part', 'Exercise_Name', 'Equipment_Type',]
)['Total_Reps'].diff()
```

```python
train_df['progress_rate'] = (
    train_df['Set4_Weight'] - train_df['Initial_Weight_kg']
) / train_df.groupby('Exercise_Name')['Initial_Weight_kg'].transform('first')
```

### Calculate VIF

```python
import numpy as np
```

```python

from statsmodels.stats.outliers_influence import variance_inflation_factor
def calculate_vif(df):
    vif_df=pd.DataFrame()
    vif_df.replace([np.inf, -np.inf], np.nan, inplace=True)
    vif_df.fillna(0,inplace=True)
    vif_df['Feature']=df.columns
    vif_df['VIF']=[variance_inflation_factor(df.values,i) for i in range(df.shape[1])]
    return vif_df
```
